```python
import numpy as np
import tensorflow as tf
```

```python
np.random.seed(42)
input_data = np.random.rand(3, 4).astype(np.float32) * 10

print("--- Original Input Values ---")
print(input_data)
```

```python
batch_norm_layer = tf.keras.layers.BatchNormalization()
bn_output = batch_norm_layer(input_data, training=True)

print("--- Batch Normalization Output ---")
print(bn_output.numpy())
```

```python
layer_norm_layer = tf.keras.layers.LayerNormalization()
ln_output = layer_norm_layer(input_data)

print("--- Layer Normalization Output ---")
print(ln_output.numpy())
```

```python
print("What Batch Normalization does:")
print("- Normalizes inputs across the batch dimension for each feature independently, stabilizing and speeding up training.")

print("\nWhat Layer Normalization does:")
print("- Normalizes inputs across the feature dimension for each individual example, useful for sequence models.")

print("\nKey Difference:")
print("- BatchNorm relies on batch statistics (mean/variance across samples), so it struggles with very small batches.")
print("- LayerNorm computes statistics per sample, making it independent of batch size.")
```
